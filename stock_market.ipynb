# ***Exploratory Data Analysis***

```python
# first we are counting the person based on gender
plt.figure(figsize=(3,5))
gen = sns.countplot( x="Gender",data = sale)
gen.bar_label(gen.containers[0])
```

*From the about data we can see that there are more female then males*

```python
# ploting bar chart for amount spend on gender base
sale_gen = sale.groupby(['Gender'],as_index=False)["Amount"].sum()  #we use as_index =False

sns.barplot(x = 'Gender',y= 'Amount' ,data = sale_gen)

plt.show()
```

**The amount spend by females is more then the males**

```python
neww_= sns.countplot(data = sale, x = 'Age Group', hue = 'Gender')
neww_.bar_label(neww_.containers[0])
```

```python
final = sale.groupby(["Age Group"],as_index = False)['Amount'].sum()
sns.barplot(data =final , x = "Age Group"  , y = 'Amount')
```

We can se that most of the people who spend money are females and most of them are of age group 26-35

```python
# state and number of order

ste = sale.groupby(["State"],as_index = False)["Orders"].sum().sort_values(by = "Orders" ,ascending= False).head(10)
sns.set(rc={'figure.figsize':(17,4)})
x = sns.barplot(data = ste, x = 'State',y= 'Orders')



for p in x.patches:
    x.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.show()
```

```python
# state and amount paid


ste = sale.groupby(["State"],as_index = False)["Amount"].sum().sort_values(by = "Amount" ,ascending= False).head(10)
sns.set(rc={'figure.figsize':(17,4)})
sns.barplot(data = ste, x = 'State',y= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively *

```python

plt.figure(figsize=((4,4)))
status =sns.countplot(x= "Marital_Status" , data= sale)
status.bar_label(status.containers[0])
```

```python
newww = sale.groupby(["Marital_Status","Gender" ],as_index = False) ['Amount'].sum()
plt.figure(figsize=(5,6))
sns.barplot(x="Marital_Status" , y= 'Amount' , hue = "Gender",data = newww)
```

*fom the above graph we can see most of buyers are married espasically women who have high purchasing power*

```python
# occupation
plt.figure(figsize=((20,6)))
occ = sns.countplot(x='Occupation' ,data = sale)
occ.bar_label(occ.containers[0])
plt.show()
```

```python
# occupation vs amount

plt.figure (figsize=(20,4))
new = sale.groupby(["Occupation"],as_index = False) ['Amount'].sum().sort_values("Amount" , ascending =False)
sns.barplot(data=new, x = "Occupation" , y="Amount")
```

*from above graph we can see that most of people who spend money are from IT*

```python
# Product cat
plt.figure(figsize = (30,6))
first = sns.countplot(x="Product_Category" ,data = sale)
first.bar_label(first.containers[0])
```

```python
sales_state = sale.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = sale.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
